```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('DatosCompletos.csv')

# Convertir las columnas de fechas a formato datetime
df['FECHA_COSNUMO_DESDE'] = pd.to_datetime(df['FECHA_COSNUMO_DESDE'], format='%Y%m%d', errors='coerce')
df['FECHA_CONSUMO_HASTA'] = pd.to_datetime(df['FECHA_CONSUMO_HASTA'], format='%Y%m%d', errors='coerce')

# Gráfico de Barras: Distribución de Tarifas por Departamento
plt.figure(figsize=(12, 6))
tarifa_departamento = df.groupby('DEPARTAMENTO')['TARIFA'].value_counts().unstack().fillna(0)
tarifa_departamento.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribución de Tarifas por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de Usuarios')
plt.legend(title='Tarifa', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Gráfico de Barras: Consumo Total por Departamento
consumo_por_departamento = df.groupby('DEPARTAMENTO')['CONSUMO'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=consumo_por_departamento, x='DEPARTAMENTO', y='CONSUMO', palette='coolwarm')
plt.title('Consumo Total por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Consumo Total (KWh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de Barras: Importe Promedio por Tarifa
importe_promedio_tarifa = df.groupby('TARIFA')['IMPORTE'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=importe_promedio_tarifa, x='TARIFA', y='IMPORTE', palette='pastel')
plt.title('Importe Promedio por Tarifa')
plt.xlabel('Tarifa')
plt.ylabel('Importe Promedio (S/)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de Barras: Consumo Promedio por Departamento
consumo_promedio_por_departamento = df.groupby('DEPARTAMENTO')['CONSUMO'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=consumo_promedio_por_departamento, x='DEPARTAMENTO', y='CONSUMO', palette='magma')
plt.title('Consumo Promedio por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Consumo Promedio (KWh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
