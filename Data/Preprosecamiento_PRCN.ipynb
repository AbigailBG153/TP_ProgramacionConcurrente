```python
import pandas as pd
import numpy as np
from google.colab import files

file_path = '/content/DatosCompletos.csv'
df = pd.read_csv(file_path, low_memory=False)

# Función para limpiar fechas incorrectas
def safe_to_datetime(col, format):
    return pd.to_datetime(col, format=format, errors='coerce')

# Convertir las fechas a formato datetime
df['FECHA_COSNUMO_DESDE'] = safe_to_datetime(df['FECHA_COSNUMO_DESDE'], '%Y%m%d')
df['FECHA_CONSUMO_HASTA'] = safe_to_datetime(df['FECHA_CONSUMO_HASTA'], '%Y%m%d')

# Eliminar filas con fechas inválidas si es necesario
df = df.dropna(subset=['FECHA_COSNUMO_DESDE', 'FECHA_CONSUMO_HASTA'])

# Calcular DIAS_CONSUMO
df['DIAS_CONSUMO'] = (df['FECHA_CONSUMO_HASTA'] - df['FECHA_COSNUMO_DESDE']).dt.days

# Extraer MES_CONSUMO y DIA_CONSUMO
df['MES_CONSUMO'] = df['FECHA_COSNUMO_DESDE'].dt.month
df['DIA_CONSUMO'] = df['FECHA_COSNUMO_DESDE'].dt.day

# Crear columnas para las diferentes tarifas
tarifas = ['BT2', 'BT3', 'BT4', 'BT5A', 'BT5B', 'BT5C', 'BT5D', 'BT5E', 'BT6', 'BT8',
           'MT2', 'MT2E', 'MT3', 'MT4']
for tarifa in tarifas:
    df[f'TARIFA_{tarifa}'] = np.where(df['TARIFA'] == tarifa, 1, 0)

# Crear columnas para los departamentos y provincias
departamentos = df['DEPARTAMENTO'].unique()
provincias = df['PROVINCIA'].unique()

for departamento in departamentos:
    df[f'DEPARTAMENTO_{departamento}'] = np.where(df['DEPARTAMENTO'] == departamento, 1, 0)

for provincia in provincias:
    df[f'PROVINCIA_{provincia}'] = np.where(df['PROVINCIA'] == provincia, 1, 0)

# Crear columnas para los distritos
distritos = df['DISTRITO'].unique()

for distrito in distritos:
    df[f'DISTRITO_{distrito}'] = np.where(df['DISTRITO'] == distrito, 1, 0)

# Seleccionar las columnas finales
columnas_finales = ['FECHA_COSNUMO_DESDE', 'FECHA_CONSUMO_HASTA', 'IMPORTE', 'CONSUMO', 'DIAS_CONSUMO',
                    'MES_CONSUMO', 'DIA_CONSUMO'] + \
                   [f'TARIFA_{tarifa}' for tarifa in tarifas] + \
                   [f'DEPARTAMENTO_{d}' for d in departamentos] + \
                   [f'PROVINCIA_{p}' for p in provincias] + \
                   [f'DISTRITO_{d}' for d in distritos]

df_final = df[columnas_finales]

# Guardar el archivo preprocesado
output_file = 'dataset_preprocesado.csv'
df_final.to_csv(output_file, index=False)

# Descargar el archivo
files.download(output_file)
```
